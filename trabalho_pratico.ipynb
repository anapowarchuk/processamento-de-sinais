### Importação de bibliotecas

```python
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, ifft
import numpy as np
from scipy import signal, fftpack
import warnings

warnings.filterwarnings('ignore')
```

### Importação do arquivo TXT com o ECG

```python
# Importando o Arquivo TXT
arq = open('ECG.txt', 'r')
sinal = np.loadtxt(arq) 
```

### Definição da Taxa de Amostragem como 600 Hz

```python
ts = 1/600 #taxa de amostragem
```

### Plotagem do sinal original completo

```python
t = np.arange(0, len(sinal)*ts, ts)

# Plotagem do sinal original completo
plt.rcParams['figure.figsize'] = (20, 5)
plt.subplot(1,1,1)
plt.plot(t, sinal)

plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude da Amostra")
plt.title("SINAL COMPLETO")

plt.grid()
plt.tight_layout()
plt.show()

```


### Plotagem do sinal original sem ruídos


```python
# Cortando as extremidades que apresentam ruídos
sinal_cort = sinal[int(100/ts):int(780/ts)]
t = np.arange(0, len(sinal_cort)*ts,ts)


# Plotagem do sinal original sem ruídos
plt.rcParams['figure.figsize'] = (20, 5)
plt.subplot(1,1,1)
plt.plot(t, sinal_cort)

plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude da Amostra")
plt.title("SINAL SEM RUÍDOS")

plt.grid()
plt.tight_layout()
plt.show()
```

# QUESTÃO B
Levantar o espectro de amplitude do sinal (com janelas de 3, 5 e 10 de  segundos);Comparare e discuta os resultados.

### **Espectro de amplitude do sinal -  Janela de 3 segundos** 
Janela escolhida: 160 à 163 segundos

```python
# Plotagem do espectro de amplitude do sinal: com janela de 3s

sinal_ampliado = sinal_cort[160*600:163*600]
sinal_fft = fftpack.fft(sinal_ampliado)
magnitude = np.abs(sinal_fft)
freq = fftpack.fftfreq(sinal_fft.size, d=ts)


plt.figure(figsize= (20,5));
plt.stem(freq, magnitude, '-o');
plt.xlabel("Frequência (Hz)");
plt.ylabel("Amplitude");
plt.title("ESPECTRO DE AMPLITUDE DO SINAL COMPLETO (janela de 3s)");

plt.grid()
plt.tight_layout()
plt.show()


print ("\n")


plt.figure(figsize= (20,5));
plt.xlim(left=0, right=300)
plt.stem(freq, magnitude, '-o');
plt.xlabel("Frequência (Hz)");
plt.ylabel("Amplitude");
plt.title("ESPECTRO DE AMPLITUDE DO SINAL (janela de 3s)");

plt.grid()
plt.tight_layout()
plt.show()
```

### **Espectro de amplitude do sinal -  Janela de 5 segundos** 
Janela escolhida: 160 à 165 segundos

```python
# Plotagem do espectro de amplitude do sinal: com janela de 5s

sinal_ampliado = sinal_cort[160*600:165*600]
sinal_fft = fftpack.fft(sinal_ampliado)
magnitude = np.abs(sinal_fft)
freq = fftpack.fftfreq(sinal_fft.size, d=ts)


plt.figure(figsize= (20,5))
plt.stem(freq, magnitude, '-o')
plt.xlabel("Frequência (Hz)")
plt.ylabel("Amplitude")
plt.title("ESPECTRO DE AMPLITUDE DO SINAL COMPLETO (janela de 5s)")

plt.grid()
plt.tight_layout()
plt.show()


print ("\n")


plt.figure(figsize= (20,5))
plt.xlim(left=0, right=300)
plt.stem(freq, magnitude, '-o')
plt.xlabel("Frequência (Hz)")
plt.ylabel("Amplitude")
plt.title("ESPECTRO DE AMPLITUDE DO SINAL (janela de 5s)")

plt.grid()
plt.tight_layout()
plt.show()
```

### **Espectro de amplitude do sinal -  Janela de 10 segundos** 
Janela escolhida: 160 à 170 segundos


```python
# Plotagem do espectro de amplitude do sinal: com janela de 10s

sinal_ampliado = sinal_cort[160*600:170*600]
sinal_fft = fftpack.fft(sinal_ampliado)
magnitude = np.abs(sinal_fft)
freq = fftpack.fftfreq(sinal_fft.size, d=ts)

plt.figure(figsize= (20, 5))
plt.stem(freq, magnitude, '-o')
plt.xlabel("Frequência (Hz)")
plt.ylabel("Amplitude")
plt.title("ESPECTRO DE AMPLITUDE DO SINAL COMPLETO (janela de 10s)")

plt.tight_layout()
plt.grid()
plt.show()


print ("\n")


plt.figure(figsize= (20,5))
plt.xlim(left=0, right=300)
plt.stem(freq, magnitude, '-o')
plt.xlabel("Frequência (Hz)")
plt.ylabel("Amplitude")
plt.title("ESPECTRO DE AMPLITUDE DO SINAL (janela de 10s)")

plt.tight_layout()
plt.grid()
plt.show()
```

# **QUESTÃO D**
Filtrar o sinal com a transformada de Fourier removendo o ruído detectado na questão c.

```python
fft = fftpack.fft(sinal_cort)
magnitude = np.abs(fft)
freq = fftpack.fftfreq(sinal_cort.size, d=ts)

alta_freq = fft.copy()
alta_freq[np.where(magnitude < maior_magnitude)] = 0
alta_freq[np.abs(magnitude < maior_magnitude)] = 0
resultado_filtragem = fftpack.ifft(alta_freq)
```

# **QUESTÃO E**
Plotar o sinal filtrado no domínio do tempo.

```python
# Plotando o espectro de amplitude
plt.rcParams['figure.figsize'] = (20, 15)

t = np.arange(0, len(sinal_cort)*ts, ts)

plt.subplot(3,1,1)
plt.xlim(left=200, right=210)
plt.plot(t, resultado_filtragem, 'tab:red', label="Sinal filtrado")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude do Sinal")
plt.title("SINAL FILTRADO AMPLIANDO (EM JANELA DE 10S)")
plt.legend(loc = "upper center")

plt.tight_layout()
plt.grid();
plt.show();
```

```python
plt.rcParams['figure.figsize'] = (20, 15)

t = np.arange(0, len(sinal_cort)*ts, ts)

plt.subplot(3,1,2)
plt.xlim(left=200, right=210)
plt.plot(t, sinal_cort, 'tab:green', label="Sinal com ruído")
plt.plot(t, resultado_filtragem, 'tab:red', label="Sinal filtrado")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude do Sinal")
plt.title("SOBREPOSIÇÃO: SINAL FILTRADO E SINAL COM RUÍDO (JANELA DE 10S)")
plt.legend(loc = "upper center")

plt.tight_layout()
plt.grid();
plt.show();
```

```python
plt.rcParams['figure.figsize'] = (20, 15)

t = np.arange(0, len(sinal_cort)*ts, ts)

plt.subplot(3,1,3)
plt.xlim(left=200, right=201)
plt.plot(t, sinal_cort, 'tab:green', label="Sinal com ruído")
plt.plot(t, resultado_filtragem, 'tab:red', label="Sinal filtrado")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude do Sinal")
plt.title("SOBREPOSIÇÃO: SINAL FILTRADO E SINAL COM RUÍDO (JANELA DE 1S)")
plt.legend(loc = "upper center")

plt.tight_layout()
plt.grid();
plt.show();
```
